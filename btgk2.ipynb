```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

1. Obtain the images “lena.bin” and “peppers.bin” from the dataset Each imagehas 256
× 256 pixels and each pixel has 8 bits

(a) Read and display the images

```python
lena_img = np.fromfile('lena.bin', dtype=np.uint8).reshape(256,256)
peppers_img = np.fromfile('peppers.bin', dtype=np.uint8).reshape(256,256)
plt.figure(6, figsize=(8,8))

plt.subplot(2,2,1)
plt.xticks([])
plt.yticks([])
plt.imshow(lena_img, cmap='gray')

plt.subplot(2,2,2)
plt.xticks([])
plt.yticks([])
plt.imshow(peppers_img, cmap='gray')
```

b) Define a new 256 256 image J as follows: the left half of J, i.e., the first 128
columns, should be equal to the left half of the Lena image. The right half of
J, i.e., the 129th column through the 256th column, should be equal to the right
half of the Peppers image

```python
new_image_lena = lena_img[:,:128]
new_image_pepper = peppers_img[:,128:]
print(new_image_lena.shape)
print(new_image_pepper.shape)
J = np.concatenate((new_image_lena, new_image_pepper), axis=-1)

#res = np.hstack((new_image_lena, new_image_pepper))
plt.imshow(J, cmap='gray')
```

(c) Define a new 256 × 256 image K by swapping the left and right halves of J. <br>

```python
K = np.zeros(J.shape)
K[:, :128] = J[:, 128:]
K[:, 128:] = J[:,:128]
plt.imshow(K, cmap='gray')
```

(a) Obtain the color image “lena512color.jpg” from the dataset. It is the same 
image that you used in the first two problems, except this time it is in color (each 
pixel has 24 bits) and the size is 512 512 pixels. If you read the image into a 
Python array J1, then J1(:,:,1) is the Red band, J1(:,:,2) is the Green band, 
and J1(:,:,3) is the Blue band. In each band, each pixel has 8 bits, just like
the image in the first problem. <br>

(b) Use imread to read in the image and then display it. Let’s call this image J1. <br>

```python
J1 = plt.imread('lena512color.jpg', cv2.COLOR_BGR2RGB)
print(J1.shape)
J1_r = J1[:,:,0]
J1_g = J1[:,:,1]
J1_b = J1[:,:,2]
plt.xticks([])
plt.yticks([])
plt.imshow(J1)
```

(c) Make a new color image J2 by swapping the color bands of J1 as follows. First,
just set J2 = J1 to initialize the new image with the right size. Then make the
Red band of J2 equal to the Blue band of J1, make the Green band of J2 equal
to the Red band of J1, and make the Blue band of J2 equal to the Green band of
J1.<br>

```python
J2 = J1.copy()
J2[:,:,0] = J1_b
J2[:,:,1] = J1_r
J2[:,:,2] = J1_g
 
```

(d) Display the new image and use imwrite to write it out to a JPEG file. <br>

```python
plt.imshow(J2)
plt.xticks([])
plt.yticks([])
cv2.imwrite('lena512color.jpeg', J2)
```
